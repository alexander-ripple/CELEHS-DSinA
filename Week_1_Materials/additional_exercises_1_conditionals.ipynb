## Conditionals

Take as input a month name (ie 'July'), and output the number of days in the month (Assume it is not a leap year). If the input is not a month, print "This is not a valid month name"

````python
months = {'January': 31,
          'February':28,
          'March':31,
          'April':30,
          'May':31,
          'June':30,
          'July':31,
          'August':31,
          'September':30,
          'October':31,
          'November':30,
          'December':31 }
user_month = input('Type the name of a month:  ')
if user_month in months:
    print(months[user_month])
else:
    print('This is not a valid month')
````
