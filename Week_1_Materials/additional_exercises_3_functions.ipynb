# Additional Exercises 3: Functions


Write a function that tests if a number is prime or not

Write a function that prints the first `n` rows of Pascal's triangle

Given two inputs `a` and `b`, write a function that sums the numbers up to `a` raised to the `b`th power. For example, given 3, 4 as inputs, ouput the sum of 1^4 + 2^4 + 3^4.

Write a function that returns a list of all proper divisors of the input

Write a function that returns whether a number is a perfect number. A perfect number is a number whose proper divisors sum to the number. For example, 6 is perfect since 1 + 2 + 3 = 6

Given a potentially nested list of integers, print the sum of all the integers. For example, one possible input is `[3, [2, 3], 4, [1, [5, 6]]]` and you should output `24`.

Given as input a list of integers `l` and a target integer `k`, find two numbers in `l` that sum to `k`, or return `None`

Given as input a list of integers `l` and a target integer `k`, find three numbers in `l` that sum to `k`, or return `None`

Given as input a list of integers `l` and a target integer `k`, find the index in `l` where `k` is. If `k` is not in `l`, return `None`

Given as input a sorted list of integers `l` and a target integer `k`, find the index in `l` where `k` is. If `k` is not in `l`, return `None`.

Work on this if you've finished everything else: Given 4 positive integers, test if you can make 24 using the four basic operations. If so, return the sequence of operations, otherwise return `No solution`.
