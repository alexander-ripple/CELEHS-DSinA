Write a function that returns a list of all proper divisors of the input

```python
def divisors(my_num):
    n = 0
    output = []
    while n < my_num - 1:
        n += 1
        if my_num%n == 0:
            output.append(n)
    
    return output

print(divisors(30))
```

Write a function that returns whether a number is a perfect number. A perfect number is a number whose proper divisors sum to the number. For example, 6 is perfect since 1 + 2 + 3 = 6

```python
def is_perfect_number(my_number):
    my_list = divisors(my_number)
    if sum(my_list) == my_number:
        return True
    return False
```
